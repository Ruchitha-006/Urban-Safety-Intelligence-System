```python
# Check current working directory
print("Current Path:", os.getcwd())

# Define possible paths
paths = [
    "../data/fir_text_data.csv",
    "data/fir_text_data.csv",
    "fir_text_data.csv"
]

file_path = None

# Find correct path
for path in paths:
    if os.path.exists(path):
        file_path = path
        break

if file_path is None:
    raise FileNotFoundError("❌ fir_text_data.csv not found. Check your folder!")

print("✅ Found file at:", file_path)

# Load data
df = pd.read_csv(file_path)

df.head()
```

CELL 3: CHECK DATA

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nNull values:\n", df.isnull().sum())
```

CELL 4: CLEAN DATA

```python
df = df.dropna()

df['text'] = df['text'].astype(str)
df['crime_type'] = df['crime_type'].astype(str)

df.head()
```

CELL 5: SPLIT DATA

```python
X = df['text']
y = df['crime_type']
```

CELL 6: VECTORIZATION

```python
vectorizer = TfidfVectorizer()

X_vec = vectorizer.fit_transform(X)

print("Vectorized shape:", X_vec.shape)
```

CELL 7: TRAIN MODEL

```python
model = LogisticRegression()

model.fit(X_vec, y)

print("✅ Model trained successfully")
```

CELL 8: TEST MODEL

```python
sample = ["Mobile stolen near bus stand"]

sample_vec = vectorizer.transform(sample)

prediction = model.predict(sample_vec)

print("Prediction:", prediction[0])
```

CELL 9: SAVE MODEL

```python
# Create folder if not exists
save_path = "../backend/models"
os.makedirs(save_path, exist_ok=True)

# Save model + vectorizer
pickle.dump((model, vectorizer), open(f"{save_path}/classifier.pkl", "wb"))

print("✅ Model saved successfully at:", save_path)
```
